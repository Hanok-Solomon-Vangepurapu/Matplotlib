```python
import matplotlib.pyplot as plt
%matplotlib inline  

```

```python
exp_vals = [1400,600,300,410,250]
exp_labels = ["Home Rent","Food","Phone/Internet Bill","Car ","Other Utilities"]
```

```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels,radius=2,autopct="%0.2f%%",shadow=True,explode=[0,0,1,0.1,0.2],startangle=180)
#by default radius value is 1
#"%0.2f%%" format specifer
#startingangle is rotation of pie chart by default it was 45 degress
plt.show()
```

```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels,radius=2,autopct="%0.2f%%",shadow=True,explode=[0,0,0,0.1,0.2],startangle=45)
#by default radius value is 1
#"%0.2f%%" format specifer
#startingangle is rotation of pie chart by default it was 45 degress
plt.show()
plt.savefig("piechart.png",bbox_inches="tight",pad_inches=2,transparent=True)
```
